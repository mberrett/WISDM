# HTML Parser

```python
def parser(txt_file):
    fhandle = open(txt_file, encoding="utf-8", errors="ignore")
    soup = BeautifulSoup(fhandle, 'html.parser')
    clean = soup.text.strip()
    
    return clean
```

# Save Clean Text to New Folder

for file in os.listdir(path):

        out.write(clean_text)

# Sift through Contents of Clean Text

```python
links = []
dates = []
words = []
buyers = []
sellers = []


for file in os.listdir(path_t):
    fhand = open(path_t + file, encoding="utf-8", errors="ignore")
    for i in fhand:
        i = i.rstrip()
        
        # Links
        
        l = file
        if (l != '.DS_Store') & (l not in links):
                links.append(l)
        
        # Dates
        
        d = re.match('^Date: ([^\.]+)', i)
        if d == None:
            d = None
        else:
            dates.append(d.group(1))
            
        # Buyer
        b = re.match('^Buyer: ([^\.]+)', i)
        if b == None:
            b = None
        else:
            buyers.append(b.group(1))
            
        # Seller
        s = re.match('^Seller: ([^\.]+)', i)
        if s == None:
            s = None
        else:
            sellers.append(s.group(1))    
            
            
        # Words
        
        w = re.match('((?!Seller|Buyer|Date)\w.+)', i)
        if w == None:
            w = None
        else:
            words.append(w.group(1))

            
 
```

```python
content = []
for i in words:
    
    c = re.search('^Today ([^\.]+)(\s(\w+)$)', i)
    if c == None:
            c = None
    else:
            content.append(c.group(3))  
```

```python
users = []
for i in content:
    if i not in users:
        users.append(i)
```
